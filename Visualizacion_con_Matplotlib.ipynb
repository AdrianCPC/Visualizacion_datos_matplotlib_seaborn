##Primeras Visualizaciones

```python
import pandas as pd

df = pd.read_csv('base_tiempo.csv')
df.head()
```

Transformacion de las fechas a formato correcto

```python
df['date'] = pd.to_datetime(df['fecha'])
df.info()
```

Usando los graficos de matplotlib

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(df['date'], df['temperatura']);
```

```python
plt.figure(figsize=(15,8))
plt.plot(df['date'], df['temperatura']); #temp en grados kelvin
```

```python
plt.figure(figsize=(15,8))
plt.plot(df['date'], df['temperatura']);
plt.title('Temperatura en el tiempo');
```

```python
fig = plt.figure(figsize=(15,8))
eje = fig.add_axes([0,0,1,1]);
eje.plot(df['date'], df['temperatura']);
eje.set_title('Temperatura en el tiempo');
eje.set_ylabel('Temperatura');
eje.set_xlabel('Date');
```

```python
fig = plt.figure(figsize=(15,8))
eje = fig.add_axes([0,0,1,1]);
eje.plot(df['date'], df['temperatura']);
eje.set_title('Temperatura en el tiempo', fontsize=25);
eje.set_ylabel('Temperatura', fontsize=20);
eje.set_xlabel('Date', fontsize=20);
```

```python
fig = plt.figure(figsize=(15,8))
eje = fig.add_axes([0,0,1,1]);
eje2 = fig.add_axes([0.7,0.65,0.3,0.3])
eje.grid(True)
eje.plot(df['date'], df['temperatura'], color='g');
eje.set_xlim(datetime.datetime(2014,5,1),datetime.datetime(2014,6,1))
eje.set_ylim(270,320)
eje.set_title('Temperatura en Mayo 2014', fontsize=25, pad=20);
eje.legend(['temperatura'], loc='lower right', fontsize=15);
eje.set_ylabel('Temperatura', fontsize=20);
eje.set_xlabel('Date', fontsize=20);

azul_izquierda = df['date'] < datetime.datetime(2014,5,1)
azul_derecha = df['date'] > datetime.datetime(2014,6,1)


eje2.plot(df['date'], df['temperatura'], color='g');
eje2.plot(df[azul_izquierda]['date'], df[azul_izquierda]['temperatura'], color='b')
eje2.plot(df[azul_derecha]['date'], df[azul_derecha]['temperatura'], color='b')
eje2.set_xlim(datetime.datetime(2014,1,1),datetime.datetime(2015,1,1))
eje2.set_title('Temperatura en 2014', fontsize=15);
eje2.legend(['temperatura'], loc='best', fontsize=8);
eje2.set_ylabel('Temperatura', fontsize=10);
eje2.set_xlabel('Date', fontsize=10);
eje2.grid(True)
```
